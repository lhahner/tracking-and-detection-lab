```python
import numpy as np
import open3d as o3d

bin_path = "/home/lennart/Dokumente/gau/advanced-research-training-applied-system-development/track-lab/src/mot_benchmark/train/KITTI-3D/img1/000008.bin"  # your file

# Common LiDAR .bin format: float32 x,y,z,intensity
points = np.fromfile(bin_path, dtype=np.float32).reshape(-1, 4)
xyz = points[:, :3]
intensity = points[:, 3]

# Create Open3D point cloud
pcd = o3d.geometry.PointCloud(o3d.utility.Vector3dVector(xyz))

# Optional: colorize by intensity (simple normalization)
i = intensity
i = (i - i.min()) / (i.max() - i.min() + 1e-12)
colors = np.stack([i, i, i], axis=1)  # grayscale
pcd.colors = o3d.utility.Vector3dVector(colors)

# Visualize
o3d.visualization.draw_geometries([pcd])
```
